## Задача проекта
  
  Помочь маркетологам оптимизировать маркетинговые затраты, анализируя данные о посещениях и заказах с сайта и данные о маркетинговых расходах.

## Описание проекта

Проведен исследовательский анализ данных и расчитаны и визуализированы метрики - LTV, CAC, DAU, WAU, MAU, ROMIс разбивкой на источники и платформы.
